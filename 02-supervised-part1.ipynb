```python
import pandas as pd

df = pd.read_csv("/kaggle/input/datasets/bharatgarg1234/diabetes-full/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
df.head()
```

```python
# ============================
# TASK 1 — Train/Test Splits
# Full Dataset + RBF SVM
# ============================
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

SEED = 42

# Full dataset already loaded as df
X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

split_results = []

for test_size in [0.40, 0.30, 0.20]:   # 60/40, 70/30, 80/20
    print(f"\nRunning split {int((1-test_size)*100)}/{int(test_size*100)}...")

    start_time = time.time()

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        stratify=y,
        random_state=SEED
    )

    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec  = recall_score(y_test, y_pred, zero_division=0)
    f1   = f1_score(y_test, y_pred, zero_division=0)

    end_time = time.time()
    print(f"  Done in {end_time - start_time:.2f}s")

    split_results.append((test_size, acc, prec, rec, f1))

split_df = pd.DataFrame(split_results,
                        columns=["test_size", "accuracy", "precision", "recall", "f1"])
split_df
```

```python
# ============================
# TASK 2 — K-Fold Cross Validation
# Full Dataset + RBF SVM
# ============================

import pandas as pd
import time
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

SEED = 42

X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

pipe_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
])

k_results = []
k_values = [3, 6, 9]

for k in k_values:   # K = 3 to 10
    print(f"\nRunning K={k}...")

    cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=SEED)
    fold_scores = []

    start_time = time.time()

    for fold, (train_idx, test_idx) in enumerate(cv.split(X, y), 1):
        print(f"  Fold {fold}/{k}...")

        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        pipe_svm.fit(X_train, y_train)
        y_pred = pipe_svm.predict(X_test)

        acc = accuracy_score(y_test, y_pred)
        fold_scores.append(acc)

    end_time = time.time()
    mean_acc = sum(fold_scores)/len(fold_scores)
    std_acc  = pd.Series(fold_scores).std()

    print(f"  Done K={k} in {end_time - start_time:.2f}s | mean_acc={mean_acc:.4f} | std_acc={std_acc:.4f}")

    k_results.append((k, mean_acc, std_acc))

k_df = pd.DataFrame(k_results, columns=["k", "mean_accuracy", "std_accuracy"])
k_df
```

```python
#TASK 3A
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

SEED = 42

print("Preparing full dataset...")

X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

print("Splitting data (80/20)...")

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=SEED
)

print("Building BASIC SVM pipeline...")

pipe_basic = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
])

print("Training BASIC SVM (1–3 minutes)...")

pipe_basic.fit(X_train, y_train)

print("Predicting...")

y_pred_basic = pipe_basic.predict(X_test)

print("\n=== BASIC SVM RESULTS ===")
print("Test Accuracy:", accuracy_score(y_test, y_pred_basic))
print(classification_report(y_test, y_pred_basic))
```

```python
from sklearn.metrics import confusion_matrix

print("\n=== CONFUSION MATRIX (Basic SVM) ===")
cm = confusion_matrix(y_test, y_pred_basic)
print(cm)
```

```python
#TASK 3B
from sklearn.model_selection import GridSearchCV, StratifiedKFold

print("Building TUNED SVM pipeline...")

pipe_tuned = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC())
])

# Small but valid grid (keeps runtime reasonable)
param_grid = {
    "svm__kernel": ["rbf"],
    "svm__C": [1, 10],
    "svm__gamma": ["scale", "auto"]
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=SEED)

grid = GridSearchCV(
    estimator=pipe_tuned,
    param_grid=param_grid,
    scoring="accuracy",
    cv=cv,
    n_jobs=-1,
    verbose=2   # shows progress
)

print("Running GridSearchCV (20–40 minutes)...")

grid.fit(X_train, y_train)

print("\n=== TUNED SVM RESULTS ===")
print("Best Params:", grid.best_params_)
print("Best CV Accuracy:", grid.best_score_)

best_model = grid.best_estimator_
y_pred_tuned = best_model.predict(X_test)

print("\nFinal Test Accuracy:", accuracy_score(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned))
```

```python
#TUTORIAL 4 - LOGISTIC REGRESSION
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

SEED = 21

# 1) Features and target
X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

print("Class distribution:")
print(y.value_counts(normalize=True).round(3))

# 2) Train/test split (50/50) with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    stratify=y,
    random_state=SEED
)

# 3) Logistic Regression model inside a pipeline with scaling
log_reg = LogisticRegression(
    max_iter=1000,
    random_state=SEED,
    class_weight="balanced"  # optional but helpful if classes are imbalanced
)

model = make_pipeline(
    StandardScaler(),
    log_reg
)

# 4) 10-fold stratified CV on the TRAIN set
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED)

train_scores = []   # training accuracy per fold
val_scores = []     # validation accuracy per fold
fold_results = []   # full metrics per fold

for fold, (train_idx, val_idx) in enumerate(kf.split(X_train, y_train), start=1):
    X_tr  = X_train.iloc[train_idx]
    X_val = X_train.iloc[val_idx]
    y_tr  = y_train.iloc[train_idx]
    y_val = y_train.iloc[val_idx]

    # Fit on current training fold
    model.fit(X_tr, y_tr)

    # Predictions
    y_pred_train = model.predict(X_tr)
    y_pred_val   = model.predict(X_val)

    # Training accuracy (for bias)
    train_acc = accuracy_score(y_tr, y_pred_train)
    train_scores.append(train_acc)

    # Validation metrics
    acc  = accuracy_score(y_val, y_pred_val)
    prec = precision_score(y_val, y_pred_val, zero_division=0)
    rec  = recall_score(y_val, y_pred_val, zero_division=0)
    f1   = f1_score(y_val, y_pred_val, zero_division=0)

    fold_results.append([acc, prec, rec, f1])
    val_scores.append(acc)

# 5) Per-fold results and variance
fold_df = pd.DataFrame(fold_results, columns=["Accuracy", "Precision", "Recall", "F1-score"])
variance = fold_df.var()

# 6) Fit on FULL training set and evaluate on HOLD-OUT TEST set
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)

test_acc  = accuracy_score(y_test, y_test_pred)
test_prec = precision_score(y_test, y_test_pred, zero_division=0)
test_rec  = recall_score(y_test, y_test_pred, zero_division=0)
test_f1   = f1_score(y_test, y_test_pred, zero_division=0)

# Bias: |avg training CV accuracy – test accuracy|
bias = abs(np.mean(train_scores) - test_acc)
print("Mean Train Scores (Accuracy)")
print(np.mean(train_scores))
print("Mean Test Scores (Accuracy)")
print(test_acc)

# 7) Print nicely
print("\n===== Per-fold Validation Metrics =====")
print(fold_df.round(4))

print("\n===== Validation Means (± std) =====")
print((fold_df.mean().round(4)).to_string())
print((fold_df.std().rename(lambda s: s + ' std').round(4)).to_string())

print("\n===== Variance Across Folds =====")
print(variance.round(6))

print("\n===== Test Set Performance =====")
print(f"Accuracy:  {test_acc:.4f}")
print(f"Precision: {test_prec:.4f}")
print(f"Recall:    {test_rec:.4f}")
print(f"F1-score:  {test_f1:.4f}")

print("\n===== Model Bias =====")
print(f"Bias (|avg train CV - test|): {bias:.4f}")
```
