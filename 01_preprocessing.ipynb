```python
#WEEK 3 LAB!!!!!!!!!!!!!!!
import numpy as np
import random

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)

# df is already loaded from your CSV
# Target and features:
X = df.drop(columns=["Diabetes_binary"])   # all input features
y = df["Diabetes_binary"]                  # label we want to predict
```

```python
# ==== Task 1 FULL LOOP (Subset + Linear SVM) ====

# 1) Take subset
df_small = df.sample(10000, random_state=SEED)

X = df_small.drop(columns=["Diabetes_binary"])
y = df_small["Diabetes_binary"]

split_results = []

for test_size in [0.30, 0.25, 0.20]:

    # Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        stratify=y,
        random_state=SEED
    )

    # Linear SVM pipeline
    pipe_basic = Pipeline([
        ("scaler", StandardScaler()),
        ("svm", SVC(kernel="linear", C=1.0, random_state=SEED))
    ])

    # Train
    pipe_basic.fit(X_train, y_train)

    # Predict
    y_pred = pipe_basic.predict(X_test)

    # Metrics
    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec  = recall_score(y_test, y_pred)
    f1   = f1_score(y_test, y_pred)

    split_results.append((test_size, acc, prec, rec, f1))

# Table
split_df = pd.DataFrame(split_results,
                        columns=["test_size", "accuracy", "precision", "recall", "f1"])
split_df

```

```python
# ==== Task 2 — K-Fold Cross-Validation (Subset + Linear SVM) ====

# 1) Use the same subset as Task 1
df_small = df.sample(10000, random_state=SEED)

X = df_small.drop(columns=["Diabetes_binary"])
y = df_small["Diabetes_binary"]

# 2) Define the basic SVM pipeline (linear kernel for speed)
pipe_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="linear", C=1.0, random_state=SEED))
])

# 3) Loop over K = 3 to 10
k_results = []

for k in range(3, 11):
    cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=SEED)
    scores = cross_val_score(pipe_svm, X, y, cv=cv, scoring="accuracy")
    k_results.append((k, scores.mean(), scores.std()))

# 4) Convert to table
k_df = pd.DataFrame(k_results, columns=["k", "mean_accuracy", "std_accuracy"])
k_df
```

```python
# ==== TASK 3A — BASIC SVM (FULL DATASET, SAFE VERSION) ====

print("Loading data...")

X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

print("Splitting data...")

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=SEED
)

print("Building pipeline...")

pipe_basic = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
])

print("Training BASIC SVM (this may take ~1–2 minutes)...")

pipe_basic.fit(X_train, y_train)

print("Predicting...")

y_pred_basic = pipe_basic.predict(X_test)

print("\n=== BASIC SVM RESULTS ===")
print("Test Accuracy:", accuracy_score(y_test, y_pred_basic))
print(classification_report(y_test, y_pred_basic))
```
